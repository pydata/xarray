```python
import dask.array as da
from IPython.display import HTML, display

import xarray as xr
```

```python
class TestDuckArray(da.Array):
    def _repr_html_inline_(self):
        return f"<span class='xr-var-item'>TestDuckArray of shape {self.shape}</span>"
```

```python
a = da.linspace(0, 1, 20, chunks=2)

test_array = TestDuckArray(
    dask=a.__dask_graph__(), name=a.name, chunks=a.chunks, dtype=a.dtype, meta=a._meta
)
display(HTML(test_array._repr_html_inline_()))
```
